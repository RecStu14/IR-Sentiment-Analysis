# RNN Training - Vader Labels

Training using the labels generated by the Vader Sentiment Analyzer.

```python
import matplotlib.pyplot as plt
import pandas as pd
import re
```

```python
dataset = pd.read_csv('skincare/product_reviews_labels.csv')
dataset.head()
```

```python
dataset.drop(columns=['blob_label'], axis=1)
dataset.head()
```

### Data Processing

#### Remving Unwanted Punctuations

```python
def depure_data(data):
    # Remove new line characters
    data = re.sub('\s+', ' ', data)

    # Remove distracting single quotes
    data = re.sub("\'", "", data)
        
    return data
```

#### Converting text into Lowercase

```python
for idx, row in dataset.iterrows():
    dataset.at[idx,'review'] = str(row['review']).lower()
```
